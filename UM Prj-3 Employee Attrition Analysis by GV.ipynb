### **Correlation Analysis**
The correlation matrix heatmap provides a visual representation of the Pearson correlation coefficients between different features in the dataset. Each cell in the matrix indicates the correlation between variables on the corresponding x and y axes. The color intensity reflects the strength of these correlations, with dark red highlighting strong positive correlations, dark blue indicating strong negative correlations, and lighter shades representing weaker correlations.

```python
plt.figure(figsize=(12, 8))
corr_matrix = data.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap='PuBuGn', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

### **Observation**

1. **Age and Total Working Years**:

2. **Job Level and Monthly Income**: 
   - A highly positive correlation exists between Job Level and Monthly Income. This indicates that employees at higher job levels tend to earn higher monthly incomes, highlighting the direct relationship between job hierarchy and compensation.

3. **Years at Company and Years with Current Manager**: 
   - These two features exhibit a significant positive correlation. This suggests that employees who have spent more time at the company are likely to have a longer tenure with their current manager, possibly due to managerial stability or long-term employee-manager relationships.

4. **Years Since Last Promotion and Years at Company**: 
   - A notable positive correlation is observed between these two variables, indicating that the longer an employee has been with the company, the more time has likely passed since their last promotion. This may reflect typical promotion cycles or career progression patterns within the company.

5. **Environment Satisfaction, Job Satisfaction, and Work-Life Balance**: 
   - These factors are moderately correlated with one another. This suggests that employees who are satisfied with their work environment tend to also be satisfied with their jobs and maintain a good work-life balance. The interrelationship among these factors highlights their combined impact on overall employee well-being.

6. **Low Correlation with Performance Rating**: 
   - Performance Rating shows relatively low correlation with most other features. This could imply that performance ratings are influenced by various other factors not measured here or that they do not have a direct influence on other aspects of the employee's profile. This may also suggest that performance is evaluated independently of other characteristics.

7. **Employee ID**: 
   - As expected, Employee ID shows no correlation with any other variables. This is because Employee ID serves solely as a unique identifier and does not carry any inherent data that would correlate with other features.

**Conclusion:**
The correlation matrix offers a clear understanding of the relationships between different features within the dataset. ## **Machine Learning: Splitting the data into Traning and Testing Sample**
We don't use the entire dataset to create the model. Instead, a portion of the data is randomly selected and set aside to evaluate the model's performance. This portion is known as Testing Data, while the remaining data, used to build the model, is called Training Data. Typically, 70% of the data is allocated as Training Data, and the remaining 30% is used as Testing Data.

```python
x=data.drop(['Attrition','Over18'],axis=1)
y=data['Attrition'].values
```

### *Resampling*

```python
pip install imblearn
```

```python
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
print(Counter(y))
rus=RandomOverSampler(random_state=42)
X_over,y_over = rus.fit_resample(x,y) 
print(Counter(y_over))
```

```python
# split the data into training and Testing set
from sklearn.model_selection import train_test_split
x_train,x_test, Y_train, y_test=train_test_split(X_over,y_over, test_size=0.2, random_state=42)
```

```python
# sanity check for the sampled data
print(x_train.shape)
print(x_test.shape)
print(Y_train.shape)
print(y_test.shape)
```

### **Logistic Regression**
Logistic regression is a statistical method for predicting binary classes. The outcome or target variable is dichotomous in nature. Dichotomous means there are only two possible classes. For example, it can be used for cancer detection problems. It computes the probability of an event occurrence. 
It is a special case of linear regression where the target variable is categorical in nature. It uses a log of odds as the dependent variable. Logist c Regression predicts the probability of occurrence of a binary event utilizing a logit functio

The standard cutoff is 0.5, which means that if the predicted probability is greater than 0.5, that observation is classified as a “positive” (or simply as a 1).. 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, roc_auc_score
```

```python
logreg= LogisticRegression()
logreg.fit(x_train, Y_train)
```

```python
prediction=logreg.predict(x_test)
cnf_matrix= confusion_matrix(y_test,prediction)
print("Accuracy Score-",accuracy_score(y_test, prediction))
```

```python
fig= plt. figure(figsize=(15,6))
ax1 = fig.add_subplot(1,2,1)
ax1 = sns.heatmap(pd.DataFrame(cnf_matrix), annot = True , cmap= 'Oranges' , fmt = 'd')
bottom, top = ax1.get_ylim()
ax1.set_ylim(bottom + 0.5, top - 0.5)
plt.xlabel('predicted')
plt.ylabel('expected')


ax2= fig. add_subplot(1,2,2)
y_pred_proba = logreg.predict_proba(x_test)[::1]
fpe, tpr, _ = roc_curve( y_test, prediction)
auc = roc_auc_score(y_test, prediction)
ax2 = plt.plot(fpe, tpr, label = "data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```
